```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# REMOVING UNNECCARY FEATURES AND ONE-HOT-ENCODING


```python
df = pd.read_csv('CarPrice_Assignment.csv')

df = df.drop(['car_ID', 'CarName', 'peakrpm' , 'fuelsystem' ,'enginelocation' ] , axis = 1)
#dropping peakrpm , fuelsystem due to redundancy and engine location because majority of them are located in the front 
df  = pd.get_dummies(df , columns=['fueltype', 'aspiration', 'doornumber', 'carbody', 
                                 'drivewheel', 'enginetype', 
                                 'cylindernumber'], drop_first=True)


df
```

```python
sns.histplot(df['carheight'] , kde = True)
print("carheight mean" , df['carheight'].mean())
sns.histplot(df['carwidth'] , kde = True)



```

## this is a non normal distribution about the mean , roughly all the coloumns potray the same


```python
non_hot_encoded = df.select_dtypes(include=['float64', 'int64']).columns
# the hot encoded values are of uint8 that's why it gets excluded here

for i in non_hot_encoded:
    df[i] = (df[i] - df[i].mean())/df[i].std()

df
```

# Removing outliers using Z-Score
## anything falling outside the range -3 and 3  be considered as an outlier


```python
for i in non_hot_encoded:

    outliers  = df.loc[(df[i] > 3 ) | (df[i] < -3)]

    if not outliers.empty:
        print(f'outliers in the coloumn \n {i}')
        print()
        print(outliers.loc[: , i])
    else:
        print(f'outliers not there in the coloumn : {i}')
```

## deleting these outliers from the set


```python
for i in non_hot_encoded:
    df = df.query(f"({i} <= 3) & ({i} >= -3)")
df.reset_index(drop=True, inplace=True)
```

```python

for i , column in enumerate(non_hot_encoded , 1):
    plt.subplot(4,4,i)
    sns.boxplot(df[column])
plt.tight_layout()
```

### splittting the dataset into training and testing parts


```python
#X AND Y ARE feature values and target values respectively

X = df.drop('price' , axis = 1)
Y = df['price']

#80% trainging case and 20% testing case
i = int(0.7*len(df))
X_train = X[:i]

X_test = X[i:]

Y_train  = Y[:i]
Y_test  = Y[i:]
print(X.shape,
Y.shape,
X_train.shape,
X_test.shape,
Y_train.shape,
Y_test.shape,)

    
```

```python
class Linear_Regression:

    def __init__ (self , lr = 0.01 , n_iterations  = 1000 , regularization_constant  = 1):
        self.lr = lr
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        self.regularization_constant = regularization_constant

    def fit(self, X,y):
        number_of_samples , number_of_features  = X.shape
        #starting off the gradient descent at w1,w2,...,w34 and bias at 0
        self.weights = np.zeros(number_of_features)
        self.bias = 0 
        #gradient descent for 1000 iterations
        
        for _ in range(self.n_iterations):

            y_predict  = np.dot(X , self.weights) + self.bias # we took X and self weights in that specific order since it is a matrix multiplication

            dw  = (1/number_of_samples)*(  np.dot(X.T , y_predict  - y)) + (self.regularization_constant/number_of_samples)*(self.weights)
            # here we had to take a transpose because the shape of X is 183,34 and y predict is 183, and for matrix multiplication
            # we need 34,183 * 183, so we took transpose of matrix X
            #self.regularization constant is the value of lambda

            db  = (1/number_of_samples)*np.sum(y_predict  - y )
            #updating the weights
            
            self.weights =self.weights - self.lr*dw
            self.bias= self.bias - self.lr*db

    def predict(self,X):
        predictions  = np.dot(X,self.weights) +self.bias
        return predictions


    

    

            
            
        
```

```python
class Error_checkers:

    def MSE(self , Y , Y_predicted):

        return np.mean((Y -Y_predicted)**2)
        # returns the MSE of the model

    def RMSE(self , Y , Y_predicted):
        return  (np.mean((Y -Y_predicted)**2))**0.5

    
```

```python
linearreg = Linear_Regression()
linearreg.fit(X_train , Y_train)

y_predicted  = linearreg.predict(X_test)

print(y_predicted)
checker = Error_checkers()
print("Mean Squared Error:")
print(checker.MSE(Y_test , y_predicted))

print("RMS ERROR")
print(checker.RMSE(Y_test , y_predicted))

```
